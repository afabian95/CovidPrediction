```python
from datetime import *
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
test_data = pd.read_csv(".\\Raw Data\\Testing Numbers\\global-testing.csv")

countries = test_data.Entity.unique()

country_count = 0
for country in countries:
    country_name = country.split(' - ')[0]
    country_data = test_data[(test_data == country).any(axis=1)]
    country_data.index = range(len(country_data))
    file_path = '.\\Parsed Data\\Global\\' + country_name + '.csv'
    try:
        new_data = pd.read_csv(file_path)

        all_daily_tests = []
        day = datetime(2020, 1, 1)
        first_day_index = -1
        last_day_index = -1
        increment = 0
        while day <= datetime.today():
            date_str = day.strftime("%Y-%m-%d")
            daily_tests = ''
            # Search the entire csv file only for the first two dates
            if last_day_index == -1:
                for index, row in country_data.iterrows():
                    if date_str == row['Date']:
                        daily_tests = row['Daily change in cumulative total']
                        if first_day_index == -1:
                            first_day_index = index
                        elif last_day_index == -1:
                            last_day_index = index
                            increment = last_day_index - first_day_index
            # Once the first two dates are found, continue incrementally in that direction until end of file
            else:
                last_day_index += increment
                if (last_day_index >= 0 and last_day_index < len(country_data)):
                    row = country_data.iloc[last_day_index]
                    daily_tests = row['Daily change in cumulative total']

            all_daily_tests.append(daily_tests)
            day = day + timedelta(days=1)

        new_data['daily_tests'] = all_daily_tests

        country_count += 1
        new_data.to_csv(file_path, index=False)
        print(country_name + ' (' + str(country_count) + '/' + str(len(countries)) + ') completed')
    except:
        country_count += 1
        print(country_name + ' (' + str(country_count) + '/' + str(len(countries)) + ') could not be found')
```
