```python
from datetime import *
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(".\\Raw Data\\Covid Numbers\\states-covid.csv")

states = data.state.unique()
state_count = 0
for state in states:
    state_data = data[(data == state).any(axis=1)]
    state_data.index = range(len(state_data))
    new_data = pd.DataFrame(columns = ['date', 'cases', 'deaths'])

    day = datetime(2020, 1, 1)
    first_day_index = -1
    last_day_index = -1
    increment = 0
    while day <= datetime.today():
        date_str = day.strftime("%Y-%m-%d")
        cases = ''
        deaths = ''
        # Search the entire csv file only for the first two dates
        if last_day_index == -1:
            for index, row in state_data.iterrows():
                if date_str == row['date']:
                    cases = row['cases']
                    deaths = row['deaths']
                    if first_day_index == -1:
                        first_day_index = index
                    elif last_day_index == -1:
                        last_day_index = index
                        increment = last_day_index - first_day_index
        # Once the first two dates are found, continue incrementally in that direction until end of file
        else:
            last_day_index += increment
            if (last_day_index >= 0 and last_day_index < len(state_data)):
                row = state_data.iloc[last_day_index]
                cases = row['cases']
                deaths = row['deaths']
                    
        new_data = new_data.append({'date': date_str, 'cases': cases, 'deaths': deaths}, ignore_index=True)
        day = day + timedelta(days=1)

        new_data.to_csv('.\\Parsed Data\\States\\' + state + '.csv', index=False)
    
    state_count += 1
    print(state + ' (' + str(state_count) + '/' + str(len(states)) + ') completed')
```
