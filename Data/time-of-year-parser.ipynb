```python
from os import walk
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
files_locations = [".\\Parsed Data\\Global\\", ".\\Parsed Data\\States\\"]
```

```python
for files_location in files_locations:
    files = []
    for (dirpath, dirnames, filenames) in walk(files_location):
        files.extend(filenames)
        break

    file_count = 0
    for filename in files:
    
        data = pd.read_csv(files_location + filename)

        time_index = []
        time_of_year = 0
        for i in range(len(data)):
            time_index.append(time_of_year)
            if i < 365:
                time_of_year += 1/365
            elif i > 365:
                time_of_year += 1/364
            else:
                time_of_year = 0

        data['time_of_year'] = time_index

        data.to_csv(files_location + filename, index=False)

        file_count += 1
        print(str(filename) + ' (' + str(file_count) + '/' + str(len(files)) + ') completed')
```
