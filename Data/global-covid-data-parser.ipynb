```python
from datetime import *
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(".\\Raw Data\\Covid Numbers\\global-covid-data.csv")

countries = data.Country.unique()
country_count = 0
for country in countries:
    country_data = data[(data == country).any(axis=1)]
    country_data.index = range(len(country_data))
    new_data = pd.DataFrame(columns = ['date', 'cases', 'deaths'])

    day = datetime(2020, 1, 1)
    first_day_index = -1
    last_day_index = -1
    increment = 0
    while day < datetime.now():
        date_str = day.strftime("%Y-%m-%d")
        cases = ''
        deaths = ''
        # Search the entire csv file only for the first two dates
        if last_day_index == -1:
            for index, row in country_data.iterrows():
                if date_str == row['Date_reported']:
                    cases = row['Cumulative_cases']
                    deaths = row['Cumulative_deaths']
                    if first_day_index == -1:
                        first_day_index = index
                    elif last_day_index == -1:
                        last_day_index = index
                        increment = last_day_index - first_day_index
        # Once the first two dates are found, continue incrementally in that direction until end of file
        else:
            last_day_index += increment
            if (last_day_index >= 0 and last_day_index < len(country_data)):
                row = country_data.iloc[last_day_index]
                cases = row['Cumulative_cases']
                deaths = row['Cumulative_deaths']
                        
        new_data = new_data.append({'date': date_str, 'cases': cases, 'deaths': deaths}, ignore_index=True)
        day = day + timedelta(days=1)

        new_data.to_csv('.\\Parsed Data\\' + country + '.csv', index=False)
    
    country_count += 1
    print(str(country) + ' (' + str(country_count) + '/' + str(len(countries)) + ') completed')
```
