```python
from os import walk
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
files_locations = [".\\Parsed Data\\Global\\", ".\\Parsed Data\\States\\"]
```

```python
def known(num):
    return num == num and num != ''

def percent_change(future_new, today_new, today_total):
    change = 0
    if today_new == 0 and future_new == 0:
        return change
    if today_total == 0:
        change = (future_new - today_new + 1)/(today_total + 1)
    else:
        change = (future_new - today_new)/today_total
    if change > 1:
        return 1
    if change < -1:
        return -1
    return change

for files_location in files_locations:
    files = []
    for (dirpath, dirnames, filenames) in walk(files_location):
        files.extend(filenames)
        break

    file_count = 0
    for filename in files:
        data = pd.read_csv(files_location + filename)

        # calculate daily changes in cases and deaths
        case_changes = []
        death_changes = []

        for i in range(len(data)):
            new_cases = ''
            new_deaths = ''
            change_cases = ''
            change_deaths = ''
            today = data.iloc[i]
            yesterday = data.iloc[i-1]

            if known(today['cases']):
                if known(yesterday['cases']):
                    new_cases = today['cases'] - yesterday['cases']
                else:
                    new_cases = today['cases']

            if known(today['deaths']):
                if known(yesterday['deaths']):
                    new_deaths = today['deaths'] - yesterday['deaths']
                else:
                    new_deaths = today['deaths']
            case_changes.append(new_cases)
            death_changes.append(new_deaths)


        # calculate % changes in cases and deaths
        w1_changes = [[], []]
        w2_changes = [[], []]

        for i in range(len(data)):
            w1_change_cases = ''
            w1_change_deaths = ''
            w2_change_cases = ''
            w2_change_deaths = ''
            today_new_cases = case_changes[i]
            today_new_deaths = death_changes[i]
            today_total_cases = data.iloc[i]['cases']
            today_total_deaths = data.iloc[i]['deaths']

            if i + 7 < len(data):
                next1w_new_cases = case_changes[i + 7]
                if known(today_new_cases) and known(next1w_new_cases):
                    w1_change_cases = percent_change(next1w_new_cases, today_new_cases, today_total_cases)
                next1w_new_deaths = death_changes[i + 7]
                if known(today_new_deaths) and known(next1w_new_deaths):
                    w1_change_deaths = percent_change(next1w_new_deaths, today_new_deaths, today_total_deaths)

            if i + 14 < len(data):
                next2w_new_cases = case_changes[i + 14]
                if known(today_new_cases) and known(next2w_new_cases):
                    w2_change_cases = percent_change(next2w_new_cases, today_new_cases, today_total_cases)    
                next2w_new_deaths = death_changes[i + 14]
                if known(today_new_deaths) and known(next2w_new_deaths):
                    w2_change_deaths = percent_change(next2w_new_deaths, today_new_deaths, today_total_deaths)

            w1_changes[0].append(w1_change_cases)
            w1_changes[1].append(w1_change_deaths)
            w2_changes[0].append(w2_change_cases)
            w2_changes[1].append(w2_change_deaths)

        data['w1_change_cases'] = w1_changes[0]
        data['w1_change_deaths'] = w1_changes[1]
        data['w2_change_cases'] = w2_changes[0]
        data['w2_change_deaths'] = w2_changes[1]

        data.to_csv(files_location + filename, index=False)

        file_count += 1
        print(str(filename) + ' (' + str(file_count) + '/' + str(len(files)) + ') completed')
```
