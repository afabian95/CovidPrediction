```python
from os import walk
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
files_location = ".\\Parsed Data\\"
```

```python
def update_col(old_dataset, new_dataset, col_name):
    if col_name in old_dataset:
        new_dataset[col_name] = old_dataset[col_name]
    else:
        new_dataset[col_name] = ''
    return new_dataset

files = []
for (dirpath, dirnames, filenames) in walk(files_location):
    files.extend(filenames)
    break

file_count = 0
for filename in files:
    data = pd.read_csv(files_location + filename)
    
    new_data = pd.DataFrame()
    col_titles = ['date', 'cases', 'deaths', 'daily_tests', 'lowest_temperature', 'highest_temperature', 'days_since_last_holiday',
                  'people_fully_vaccinated_per_hundred', 'daily_vaccinations_per_million', 'w1_change_cases', 'w1_change_deaths', 'w2_change_cases', 'w2_change_deaths']
    
    for title in col_titles:
        new_data = update_col(data, new_data, title)

    new_data.to_csv(files_location + filename, index=False)
    
    file_count += 1
    print(str(filename) + ' (' + str(file_count) + '/' + str(len(files)) + ') completed')
```
